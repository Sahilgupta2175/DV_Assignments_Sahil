```python
fig,ax=plt.subplots(figsize=(12,8))
sns.heatmap(wine_quality.corr(),cmap='viridis',annot=True)
plt.xticks(rotation=45)
```

```python
sns.boxplot(data=wine_quality,x='quality',y='alcohol')
```

```python
plt.figure(figsize=(5,5))
for i in wine_quality.columns:
    sns.boxplot(data=wine_quality[i])
    plt.title(i)
    plt.show()
```

```python
for i in wine_quality.columns:
    sns.distplot(wine_quality[i])
    plt.title(i)
    plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(wine_quality['density'],wine_quality['residual sugar'])
plt.xlabel('Density')
plt.ylabel('Residual Sugar')
plt.title('Graph B/W Density & Residual Sugar')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(wine_quality['density'],wine_quality['alcohol'])
plt.xlabel('Density')
plt.ylabel('Alcohol')
plt.title('Graph B/W Density & Alcohol')
plt.show()
```
